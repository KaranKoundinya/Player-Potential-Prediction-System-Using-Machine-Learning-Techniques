**How does the Body Mass Index (BMI) of football players correlate with their physical attributes, and how does it vary across different countries? This investigation entails calculating and visualizing the average BMI of players for each country, identifying the ten countries with both the highest and lowest average BMI scores. Furthermore, it seeks to pinpoint the player with the lowest and the player with the highest BMI in the dataset. **a) Calculate the average BMI of each country**

Body Mass Index (BMI) is a numerical value representing a person's body fat based on their weight and height

```python

df['BMI'] = df['Weight(in kg)'] / (df['Height(in cm)'] / 100) ** 2


df_bmi = df[['Full Name', 'BMI']]


print(df_bmi)
```

```python
df['BMI'].mean()
```

```python
average_bmi_per_country = df.groupby('Nationality')['BMI'].mean().reset_index()
```

```python
print(average_bmi_per_country)
```

This contains the list of all the countries in the dataset along with their average BMI

**b) Calculate and visualize 10 countries with highest average BMI**



```python
sorted_bmi = average_bmi_per_country.sort_values('BMI')
```

```python
highest_countries = sorted_bmi.tail(10)
lowest_countries = sorted_bmi.head(10)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(highest_countries['Nationality'], highest_countries['BMI'])
plt.title('Top 10 countries with Highest average BMI')
plt.xlabel('Country')
plt.ylabel('Average BMI')
plt.xticks(rotation=45)
for i, v in enumerate(highest_countries['BMI']):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom')
plt.show()
```

Advantages of selecting players from these countries:

**Physical Strength**: Players from high average BMI countries might possess greater natural strength

**Physical Presence**: These players can create a formidable presence on the field, influencing opponents' tactics and creating space for teammates

Disadvantages of selecting players from these countries

**Fitness Requirements** : Maintaining fitness levels and weight management could be more challenging for players with naturally higher BMI


**c) Calculate and visualize 10 countries with lowest average BMI**

```python
plt.figure(figsize=(10, 6))
plt.bar(lowest_countries['Nationality'], lowest_countries['BMI'])
plt.title('Lowest Average BMI by Country')
plt.xlabel('Country')
plt.ylabel('Average BMI')
plt.xticks(rotation=45)
for i, v in enumerate(lowest_countries['BMI']):
    plt.text(i, v, str(round(v, 2)), ha='center', va='bottom')
plt.show()
```

Advantages of selecting players from these countries:

**Technical Skills:** Lower body mass can contribute to better ball control, dribbling, and intricate footwork

**Pressing Game**: Lighter players may be better suited for high-pressing tactics, quickly closing down opponents

Disadvantages of selecting players from these countries:

**Physical Presence**: Players might lack physical strength and presence, affecting their performance in challenges and physical contests

Players with higher BMI might be more effective in aerially distributing long balls and crosses


**d) Player with the lowest BMI**



```python
lowest_bmi_player = df[df['BMI'] == df['BMI'].min()]['Full Name'].to_string(index=False)
lowest_bmi_value = df['BMI'].min().round(2)
result1 = lowest_bmi_player + ' ' + str(lowest_bmi_value)
print("The player with the lowest BMI is :", result1)
```

**e) Player with the highest BMI**

```python
highest_bmi_player = df[df['BMI'] == df['BMI'].max()]['Full Name'].to_string(index=False)
highest_bmi_value = df.BMI.max().round(2)
result = highest_bmi_player + ' ' + str(highest_bmi_value)
print("The player with the highest BMI is :",result)
```

**a)Create a metric called as Value next year, that give's the player's expected value next year**

```python

df['Years Left'] = df['Contract Until'].apply(lambda x: int(x) - 2022 if x != '-' else 0)


df['Value Next Year'] = df['Value(in Euro)'] * (1 + df['Years Left'])


print(df[['Known As', 'Value Next Year']])
```

This metric calculates and displays an estimate of how players' values might change in the next year based on their current values and the number of years remaining in their contracts. This estimation considers the influence of contract length on a player's perceived market value

For each player, it subtracts the current year (2022) from the year their contract ends, yielding the number of years left in their contract

This 'Value Next Year' estimation considers the impact of contract length on a player's perceived market value

**b) Visualize the top 10 players with highest expected values for next year**

```python

sorted_df = df.sort_values('Value Next Year', ascending=False)

top_10_players = sorted_df.head(10)


print(top_10_players[['Full Name', 'Value Next Year']])
```

```python
top_10_players = sorted_df.head(10)


plt.figure(figsize=(14, 10))
plt.bar(top_10_players['Full Name'], top_10_players['Value Next Year'])

for i, value in enumerate(top_10_players['Value Next Year']):
    plt.text(i, value, f'{value:,}', ha='center')


plt.xticks(rotation=45)

plt.xlabel('Player')
plt.ylabel('Value Next Year')
plt.title('Top 10 Players by Value Next Year')


plt.show()
```

The depicted graph showcases the roster of the top 10 players anticipated to have the greatest value in the upcoming year.
